#### Fill in blanks with brief description of the following terms:

- **Hypothesis**
> scientific word for guess - but it also needs to be testable

- **Null Hypothesis**
> hypothesis that there is no significant correlation between variables

- **P-Value**
> the level of marginal significance within a statistical hypothesis test. tells you how likely your data could have occured under the null hypothesis. If it's low, usually under 0.05, you should reject the null hypothesis because it's very unlikely your data occured with no correlation

### Step 3 - Use Pearsonr & Spearmanr To Answer Questions About Correlation and P-Values

<h3 style='color:red;'> First, load 'salary_numeric.csv' - where education has been re-expressed as numeric variables.</h3>

```python
# load salary_numeric.csv into dataframe where education levels have been re-expressed as numeric
df2 = pd.read_csv('salary_numeric.csv')
df2.head()
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

```python
# import pearson and spearmanr methods from scipy.stats library
from scipy.stats import pearsonr as pearsonr
from scipy.stats import spearmanr as spearmanr
```

### What is the correlation between salary and education?

#### Your Hypothesis = as eduation increases salary increases

```python
# calculate correlation and pvalue of salary and education with spearmanr method
spearmanr(df2['salary'], df2['education'])
```

positive correlation value indicates positive relationship, very low pvalue indicates unlikely to be due to random chance

#### Are these results statistically significant? How do you know?

yes, pvalue < 0.05

### What is the correlation between salary and age?

#### Your Hypothesis as age increases salary increases until about 40 years old

```python
# calculate correlation and pvalue of salary and age with pearsonr method
spearmanr(df2['salary'], df2['age'])
```

positive correlation - positive relationship, low pvalue indicates unlikely to be due to random chance

#### Are these results statistically significant? How do you know? 

yes, low pvalue

### What is the correlation between salary and industry experience? 

#### Your Hypothesis = as experience increases salary increases until about 10 years

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
pearsonr(df2['salary'], df2['industry_experience'])
```

positive correlation - positive relationship, low pvalue indicates unlikely to be due to random chance

#### Are these results statistically significant? How do you know?

yes, low pvalue

#### Crosstab of average salary grouped by education and industry

```python
# use pd.crosstab() to calculate mean salary in each industry grouped by education level
import numpy as np
pd.crosstab(index=df2['industry'], columns=df2['education'], values=df2['salary'], aggfunc=np.mean).round(0)
```

#### Heatmap of crosstab of average salary grouped by education and industry

```python
# save above result to new variable
crosstab = pd.crosstab(index=df2['industry'], columns=df2['education'], values=df2['salary'], aggfunc=np.mean).round(0)

ax = sns.heatmap(crosstab, annot=True, linewidths=.5, cmap='YlGnBu', fmt='g', xticklabels=['High School', 'Some College', "Bachelor's Degree", "Master's Degree"])
```

over all industries the salary increases with education.
